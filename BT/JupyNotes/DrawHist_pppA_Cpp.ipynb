# File input

```c++
TFile *f_pp_projectile_MB = TFile::Open("../pp_rapidityloss_MB.root");
if(!f_pp_projectile_MB || f_pp_projectile_MB->IsZombie()) { printf("File error 02\n"); return; }

TFile *f_pA_projectile_MB = TFile::Open("../pA_rapidityloss_MB.root");
if(!f_pA_projectile_MB || f_pA_projectile_MB->IsZombie()) { printf("File error MB\n"); return; }

TFile *f_pA_projectile_02 = TFile::Open("../pA_rapidityloss_02.root");
if(!f_pA_projectile_02 || f_pA_projectile_02->IsZombie()) { printf("File error 02\n"); return; }
```

# dN/dy comparison (projectile only)

```c++
auto h_pp_projectile_MB_dNdy = (TH1D*) f_pp_projectile_MB->Get("h_dNdy");
if(!h_pp_projectile_MB_dNdy) { printf("Histogram ppMB not found!\n"); return; }

auto h_pA_projectile_MB_dNdy = (TH1D*) f_pA_projectile_MB->Get("h_dNdy");
if(!h_pA_projectile_MB_dNdy) { printf("Histogram pAMB not found!\n"); return; }

auto h_pA_projectile_02_dNdy = (TH1D*) f_pA_projectile_02->Get("h_dNdy");
if(!h_pA_projectile_02_dNdy) { printf("Histogram pA02 not found!\n"); return; }
```

## Normalization+Style

```c++
h_pp_projectile_MB_dNdy->Scale(1.0/h_pp_projectile_MB_dNdy->Integral());
h_pA_projectile_MB_dNdy->Scale(1.0/h_pA_projectile_MB_dNdy->Integral());
h_pA_projectile_02_dNdy->Scale(1.0/h_pA_projectile_02_dNdy->Integral());

h_pp_projectile_MB_dNdy->SetLineStyle(9);
h_pp_projectile_MB_dNdy->SetLineColor(kRed);

h_pA_projectile_02_dNdy->SetLineStyle(2);

h_pp_projectile_MB_dNdy->SetLineWidth(2);
h_pA_projectile_MB_dNdy->SetLineWidth(2);
h_pA_projectile_02_dNdy->SetLineWidth(2);
```

## Draw

```c++
h_pp_projectile_MB_dNdy->GetXaxis()->SetRangeUser(-5,5);

auto c_dNdy = new TCanvas("c1", "dNdy of Projectile", 800, 800);
h_pp_projectile_MB_dNdy->Draw("HIST");
h_pA_projectile_MB_dNdy->Draw("HIST SAME");
h_pA_projectile_02_dNdy->Draw("HIST SAME");


auto legend_dNdy = new TLegend(0.15,0.7,0.4,0.85);
legend_dNdy->AddEntry(h_pp_projectile_MB_dNdy, "pp inclusive(MB)", "l");
legend_dNdy->AddEntry(h_pA_projectile_MB_dNdy, "pA inclusive(MB)", "l");
legend_dNdy->AddEntry(h_pA_projectile_02_dNdy, "pA central(0-2fm)", "l");
legend_dNdy->Draw();
c_dNdy->Draw();
```
